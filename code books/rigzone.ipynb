```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import re
import requests
pd.set_option('max_colwidth', 80)
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.54 Safari/537.36'}
```

```python
job_link=[]
Category=[]
```

```python
def extract_job_post(): 
    try:
        for i in range(0,10):
            url_next='https://www.rigzone.com/oil/jobs/search/?sk=machine+learning&page='+str(i)
            w1=requests.get(url_next,headers)
            soups=BeautifulSoup(w1.content,'html.parser') 
            soup=soups.findAll('a',href=True)
            for item in soup:
                if item['href'].startswith('/jobs/postings/'):
                    job_link.append('https://www.rigzone.com'+item['href'])
                    Category.append('machine_learning')
    except:
        pass
    return job_link,Category
```

```python
job_link,Category=extract_job_post()
```

```python
def rigzone_extract(df):
    rigzone_df=pd.DataFrame()
    for i in range(0,len(df)):
        print(i)
        try:
            Job_Posting_link=df['Job_Posting_link'][i]
            # Job_Posting_link='https://www.rigzone.com/oil/jobs/postings/1133414_machine_learning_engineer?s=8'
            # print(Job_Posting_link)
            Category=df['Category'][i]
            w2=requests.get(Job_Posting_link,headers)
            # print(w2.status_code)
            soups1=BeautifulSoup(w2.content,'html.parser') 
            Job_holder=soups1.find(class_='holder')
            Job_title=Job_holder.find('a').text
            # print(Job_title)
            content=soups1.find(class_='job-detail').findAll('span')
            Info=[]
            for c in content:
                Info.append(c.text)
            for ele in Info:
                if ele.strip().startswith('Education'):
                    Education=ele.replace('Education:','')
                if ele.strip().startswith('Skills'):
                    Skills=ele.replace('Skills:','')
                if ele.strip().startswith('Experience'):
                    Experience=ele.replace('Experience:','')
            try:
                if Education:
                    pass
            except:
                Education=None
            try:
                if Skills:
                    pass
            except:
                Skills=None
            try:
                if Experience:
                    pass
            except:
                Experience=None
            Details=[]
            for i in Job_holder.find('address'):
                # print(i.text)
                if i.text:
                    Details.append(i.text.replace('\n','').strip())
                    # Details.append(i.text)
            for ii in Details:
                if ii =='':
                    Details.remove(ii)
            Company_name=Details[0]
            # print(Company_name)
            try:
                Employer_type=Details[1]
            except:
                Employer_type=None
            # print(Employer_type)
            try:
                Location=Details[2]
            except:
                Location=None
            # print(Location)
            try:
                
                Descrip=soups1.find_all(class_='text')
                Description=Descrip[-1].text.replace('\n','').strip()
            except:
                Description=None
            # print(Description)
            rigzone1=pd.DataFrame({'Category':Category,'Job_Posting_link':Job_Posting_link,'Job_title':Job_title,'Company_name':Company_name,'Employer_type':Employer_type,'Location':Location,'Experience':Experience,'Education':Education,'Skills':Skills,'Description':[Description]},index=[0])
            # print(simplyhired1)
            rigzone_df=rigzone_df.append(rigzone1)
            # print(simplyhired_df)
        except Exception as e:
            print(e)
    return rigzone_df
```
