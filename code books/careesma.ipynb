```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options as FirefoxOptions
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import pandas as pd
import re
import requests
```

```python
job_link = []
Category = []
def careesma_extractor(url, category):#this functions extracts the job links from the particular websie
    try:
        for i in range(0,10): # iterates upto 10 pages
            url_next=url+str(i)
            w1=requests.get(url_next) # to access the url
            soups=BeautifulSoup(w1.content,'html.parser') 
            soup = soups.findAll(class_="JobCard-titleLink", href=True) #find all the data from html of respective class
            for item in soup: # iterates through the items which has href links
                if item['href'].startswith('/viewjob/'):
                    job_link.append('https://www.careesma.in'+item['href']) # adds the iterted data to the empty list
                    Category.append(category)
    except:
        pass         
    return job_link,Category
```

```python
job_link,Category=careesma_extractor('https://www.careesma.in/jobs?q=data+science&pg=', 'Data Science') # calls the funtion
```

```python
job_link,Category=careesma_extractor('https://www.careesma.in/jobs?q=machine+learning&pg=', 'Machine Learning')
```

```python
Job_posting_df=pd.DataFrame({"Category":Category,"Job_Posting_link":job_link})
```

```python
ds = pd.read_csv("Data_science_careesma_links.csv")
ml = pd.read_csv('Machine_learning_careesma_links.csv')
```

```python
def careesma_details_extractor(df): # this function extracts the data from the particular job link
    careesma_df = pd.DataFrame()
    for i in range(0,len(df)): #it iterates through the links upto the extracted links length
        try:
            Job_Posting_link=df['Job_Posting_link'][i] #iterates through the links in the dataframe
            Category=df['Category'][i]
            w2=requests.get(Job_Posting_link)
            soups1=BeautifulSoup(w2.content,'html.parser') 
            try:
                Location=soups1.find(class_='ViewJobHeader-property').text #extracts the text data from the respective class
                pass
            except:
                Location
            try:
                salary = soups1.find(class_= 'salary col-xs-12 col-sm-6 col-md-6 col-lg-6').text
                pass
            except:
                salary = None
            try:
                education = soups1.find(class_= 'salary col-xs-12 col-sm-6 col-md-6 col-lg-6').text
                pass
            except:
                education = None
            try:
                company = soups1.find(class_ = 'ViewJobHeader-company').text
                pass
            except:
                company=None
            try:
                title = soups1.find(class_ = 'ViewJobHeader-title').text
                pass
            except:
                title=None
            try:
                description = soups1.find(class_='viewjob-description-tab').text
                pass
            except:
                description=None
            careesma=pd.DataFrame({'Category':Category,'Job_Posting_link':Job_Posting_link,'Job_title':title,'Company_name':company,'Education':education,'Salary':salary, 'Location':Location,'Description':[description]},index=[0])
            careesma_df=careesma_df.append(careesma)
        except Exception as e: # raises an exception if something goes wrong in the time of extraction
            print(e)   
    return careesma_df
```
